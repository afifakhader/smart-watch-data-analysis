```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
df = pd.read_csv("dailyActivity_merged.csv")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
#activity date colum to date time format
```

```python
df.ActivityDate= pd.to_datetime(df.ActivityDate)
```

```python
df.dtypes
```

```python
#check for any null values
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["TotalMinutes"]= df.VeryActiveMinutes + df.FairlyActiveMinutes + df.LightlyActiveMinutes +  df.SedentaryMinutes
```

```python
df.head(1)
```

```python
df = df.drop("Id",axis = 1)
```

```python
df.describe()
```

```python
label = ["Very Active Minutes", "Fairly Active Minutes", 
         "Lightly Active Minutes", "Inactive Minutes"]
counts = df[["VeryActiveMinutes", "FairlyActiveMinutes", 
               "LightlyActiveMinutes", "SedentaryMinutes"]].mean()
colors = ['gold','lightgreen', "pink", "blue"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Total Active Minutes')
fig.update_traces(hoverinfo='percent', textinfo='percent', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
# 81.3% of Total inactive minutes in a day
# 15.8% of Lightly active minutes in a day
# On an average, only 21 minutes (1.74%) were very active
# and 1.11% (13 minutes) of fairly active minutes in a day
```

```python
df["Day"] = df["ActivityDate"].dt.day_name()
```

```python
df
```

```python
#sedentary minutes in each day of the week
```

```python
day = df["Day"].value_counts()
label = day.index
counts = df["SedentaryMinutes"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Inactive Minutes Daily')
fig.update_traces(hoverinfo='percent+label', textinfo='percent', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
#so thursday is the most inactive day of the week according to the dataset
```

```python
day = df["Day"].value_counts()
label = day.index
counts = df["VeryActiveMinutes"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='very active Minutes Daily')
fig.update_traces(hoverinfo='percent+label', textinfo='percent', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
#sunday is the day of the week in which people are most active
```

```python
#calorie analysis
calories = df["Day"].value_counts()
label = calories.index
counts = df["Calories"]
colors = ['gold','lightgreen', "pink", "blue", "skyblue", "cyan", "orange"]

fig = go.Figure(data=[go.Pie(labels=label, values=counts)])
fig.update_layout(title_text='Calories Burned Daily')
fig.update_traces(hoverinfo='label+percent', textinfo='percent', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.show()
```

```python
#most calories are burnt on tuesday 
# least number of calories are burnt on sunday
```
